```python
tech_list = ['LanguageHaveWorkedWith', 'LanguageWantToWorkWith',
       'DatabaseHaveWorkedWith', 'DatabaseWantToWorkWith',
       'PlatformHaveWorkedWith', 'PlatformWantToWorkWith',
       'WebframeHaveWorkedWith', 'WebframeWantToWorkWith',
       'MiscTechHaveWorkedWith', 'MiscTechWantToWorkWith',
       'ToolsTechHaveWorkedWith', 'ToolsTechWantToWorkWith',
       'NEWCollabToolsHaveWorkedWith', 'NEWCollabToolsWantToWorkWith', 'OpSys']
```

```python
survey_df.shape[0]
```

# wrangling algorithm, not good, have revised it in the next notebook

```python
df = pd.DataFrame(columns=(['respondent'] + tech_list))
for i in tqdm(range(survey_df.shape[0])):
    row = survey_df.iloc[i]
    row_tech_list = []
    for tech in tech_list:
        if isinstance(row[tech],str):
            row_tech_list.append(row[tech].split(';'))
        else:
            row_tech_list.append([])

    while row_tech_list != [[]] * len(row_tech_list):
        pd_row = [row['ResponseId']]
        for tl in row_tech_list:
            if tl:
                tech = tl.pop()
                pd_row.append(tech)
            else:
                pd_row.append('null')
    #     print(pd_row)
        dict_temp = dict()
        for i,col in enumerate(['respondent'] + tech_list):
            dict_temp[col] = pd_row[i]
    #     print(dict_temp)
        df = df.append(dict_temp, ignore_index=True)
    
    
```
